**TASK 1: Problem Definition & Dataset Selection**

**Problem Statement**

The objective of this project is to analyze employment and wage trends across occupations in the United States using Occupational Employment and Wage Statistics (OEWS) data. The study focuses on understanding how employment levels, wage distributions, and occupational demand have evolved over time across different industries and regions.


**Dataset Details**

The name of the selected data is Occupational Employment and Wage Statistics (OEWS).
The source of the data is Data.gov (US). This dataset consist of 500+ records & 10+ attributes. The Domian of this dataset is Human Resource.

```python
#Importing Libraries & Loading Dataset
import pandas as pd

file_path = 'oews-2009-2025_original.csv'
df = pd.read_csv(file_path)
```

```python
# Intial Overview

# Number of rows and columns
print(f"Rows: {df.shape[0]}, Columns: {df.shape[1]}")

# Data types of each column
print(df.dtypes)

# Initial Observation:
print("\n--- First 5 Rows of Data ---")
display(df.head())

# Summary Information: info()
df.info()

# Statistical Summary: describe()
display(df.describe())
```

**TASK 2: Data Cleaning & Pre-processing**

This section covers:  
- Handled missing values using median (numerical) and mode (categorical) methods
- Removed duplicate records to ensure data integrity  
- Standardized column names for consistency  
- Corrected data types for numerical and time-based columns  
- Created derived variables to support deeper analysis  


```python
# Check missing values in each column
missing_values = df.isnull().sum()

print("Missing values per column:")
display(missing_values[missing_values > 0])
```

```python
# Check missing values BEFORE
print("Missing values before imputation:")
display(df.isnull().sum()[df.isnull().sum() > 0])

# Numerical columns
num_cols = df.select_dtypes(include=['int64', 'float64']).columns
df[num_cols] = df[num_cols].fillna(df[num_cols].median())

# Categorical columns
cat_cols = df.select_dtypes(include=['object']).columns
for col in cat_cols:
    df[col] = df[col].fillna(df[col].mode()[0])

# Recalculate missing values AFTER
print("Missing values after imputation:")
display(df.isnull().sum()[df.isnull().sum() > 0])
```

```python
# Checked for duplicate rows, None was found
print("Number of duplicate rows:", df.duplicated().sum())
```

```python
# Standardized column names
df.columns = (
    df.columns
    .str.strip()
    .str.lower()
    .str.replace(' ', '_')
)

print("Updated column names:")
print(df.columns)
```

```python
# Converted data types & validated them

df['area_type'] = df['area_type'].astype('category')
df['wage_type'] = df['wage_type'].astype('category')

df['area_type'].value_counts()
df['wage_type'].value_counts()

df.dtypes

```

```python
# Created a new column as annual mean wage
df["annual_mean_wage"] = df["mean_wage"]

# Converted hourly wages to annual wages
df.loc[df["wage_type"] == "Hourly wage", "annual_mean_wage"] = (
    df["mean_wage"] * 40 * 52
)

display(df.head())
```

```python
# Created a new column on wage classification & coverted the type to category

df['annual_mean_wage'].describe()

q1 = df['annual_mean_wage'].quantile(0.25)
q2 = df['annual_mean_wage'].quantile(0.50)
q3 = df['annual_mean_wage'].quantile(0.75)

print(q1, q2, q3)

def classify_wage(wage):
    if wage < q1:
        return 'Low'
    elif wage < q2:
        return 'Medium'
    elif wage < q3:
        return 'High'
    else:
        return 'Very High'

df['wage_level_classification'] = df['annual_mean_wage'].apply(classify_wage)
df['wage_level_classification'] = df['wage_level_classification'].astype('category')
df['wage_level_classification'].value_counts()

```

```python
def classify_occupation(title):
    title = title.lower()

    if 'engineer' in title:
        return 'Engineering'
    elif 'developer' in title or 'programmer' in title:
        return 'IT'
    elif 'analyst' in title:
        return 'Analytics'
    elif 'manager' in title or 'director' in title:
        return 'Management'
    elif 'supervisor' in title or 'foreman' in title:
        return 'Supervisory'
    elif 'technician' in title:
        return 'Technical'
    elif 'assistant' in title or 'support' in title:
        return 'Support'
    elif 'clerk' in title or 'secretary' in title:
        return 'Clerical'
    elif 'officer' in title or 'executive' in title:
        return 'Administration'
    elif 'teacher' in title or 'professor' in title:
        return 'Education'
    elif 'nurse' in title or 'physician' in title:
        return 'Healthcare'
    elif 'scientist' in title or 'researcher' in title:
        return 'Research'
    elif 'operator' in title or 'machinist' in title:
        return 'Operations'
    elif 'inspector' in title or 'auditor' in title:
        return 'Compliance'
    else:
        return 'Other'

df['occupation_category'] = df['occupational_title'].apply(classify_occupation)
df['occupation_category'] = df['occupation_category'].astype('category')
df['occupation_category'].value_counts()
```

```python
# Final Preporocessing Check done
df.info()
display(df.head())
```

**TASK 3: Exploratory Data Analysis & Visualizations**

Will do Univariate, Bivariate & Multivariate Analysis



```python
# Distribution of Annual Mean Wage

import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8,5))
sns.histplot(
    df['annual_mean_wage'],
    bins=30,
    kde=True,
    log_scale=True
)
plt.title("Distribution of Annual Mean Wage (Log Scale)")
plt.xlabel("Annual Mean Wage (Log Scale)")
plt.ylabel("Frequency")
plt.show()



```

```python
# Wage Level Distribution

plt.figure(figsize=(6,4))
ax = sns.countplot(x='wage_level_classification', data=df)
plt.title("Distribution of Wage Levels (Quantile-Based)")
plt.xlabel("Wage Level")
plt.ylabel("Count")

# Add percentage labels
total = len(df)
for p in ax.patches:
    percentage = f'{100 * p.get_height() / total:.1f}%'
    ax.annotate(
        percentage,
        (p.get_x() + p.get_width() / 2., p.get_height()),
        ha='center', va='bottom'
    )

plt.show()

#Equal percentages occur because wage levels were defined using quartiles, which evenly split the data into four groups for relative wage comparison.



```

```python
# Boxplot of Annual Mean Wage

plt.figure(figsize=(10,5))
sns.boxplot(
    x='year',
    y='annual_mean_wage',
    data=df
)
plt.yscale('log')
plt.title("Annual Mean Wage Distribution by Year (Log Scale)")
plt.xlabel("Year")
plt.ylabel("Annual Mean Wage (Log Scale)")
plt.show()

#The annual mean wage distribution is highly right-skewed due to the presence of extremely high-paying roles.
#Applying a logarithmic scale improves visibility of the median and interquartile range, allowing clearer comparison of wage distributions across years.

```

```python
# Employment Distribution

top_categories = (
    df.groupby('occupation_category')['number_of_employed']
    .sum()
    .sort_values(ascending=False)
)

plt.figure(figsize=(8,5))
top_categories.plot(kind='barh')

plt.xscale('log')
plt.title("Total Employment by Occupation Category")
plt.xlabel("Total Number of Employed (Log Scale)")
plt.ylabel("Occupation Category")
plt.gca().invert_yaxis()
plt.show()


```

```python
#Average Wage by Occupation Category


plt.figure(figsize=(10,5))

sns.pointplot(
    x='occupation_category',
    y='annual_mean_wage',
    data=df,
    estimator='median',
    errorbar=None
)

plt.yscale('log')
plt.title("Median Wage by Occupation Category")
plt.xlabel("Occupation Category")
plt.ylabel("Median Annual Wage (Log Scale)")
plt.xticks(rotation=45)
plt.show()

```

```python
#Top 10 Highest Paying Area

top_states = (
    df.groupby('area_name')['annual_mean_wage']
    .mean()
    .sort_values(ascending=False)
    .head(10)
)

plt.figure(figsize=(8,5))
top_states.plot(kind='barh')
plt.title("Top 10 Highest Paying Areas")
plt.xlabel("Average Annual Wage")
plt.ylabel("Area")
plt.gca().invert_yaxis()
plt.show()
```

```python
# Occupation Category vs Number of Employees

plt.figure(figsize=(10,5))

sns.barplot(
    x='occupation_category',
    y='number_of_employed',
    data=df,
    estimator='mean'
)

plt.yscale('log')  # IMPORTANT for skewed employment data
plt.title("Number of Employed by Occupation Category")
plt.xlabel("Occupation Category")
plt.ylabel("Number of Employed (Log Scale)")
plt.xticks(rotation=45)
plt.show()


```

```python
# Wage Level vs Occupation

plt.figure(figsize=(10,5))
sns.countplot(
    x='occupation_category',
    hue='wage_level_classification',
    data=df
)
plt.title("Wage Level Distribution by Occupation")
plt.xticks(rotation=45)
plt.show()
```

```python
#Occupation  State  Wage

pivot_table = df.pivot_table(
    values='annual_mean_wage',
    index='year',
    columns='wage_level_classification',
    aggfunc='mean'
)

plt.figure(figsize=(8,5))
sns.heatmap(pivot_table, annot=True, fmt=".0f", cmap="YlGnBu")
plt.title("Year vs Wage Level Heatmap")
plt.show()
```

```python
#Year-wise Wage Distribution

plt.figure(figsize=(18,5))

sns.scatterplot(
    x='year',
    y='annual_mean_wage',
    hue='wage_level_classification',
    data=df,
    alpha=0.6,
    s=60
)

plt.yscale('log')
plt.title("Annual Mean Wage by Year and Wage Level (Log Scale)")
plt.xlabel("Year")
plt.ylabel("Annual Mean Wage (Log Scale)")
plt.legend(title="Wage Level")
plt.show()

```

```python
plt.figure(figsize=(18,5))

sns.scatterplot(
    x='occupation_category',
    y='annual_mean_wage',
    hue='wage_level_classification',
    data=df,
    alpha=0.6,
    s=60
)

plt.yscale('log')
plt.title("Annual Mean Wage by Year and Wage Level (Log Scale)")
plt.xlabel("Category")
plt.ylabel("Annual Mean Wage (Log Scale)")
plt.legend(title="Wage Level")
plt.show()
```

```python
# Area-wise wage distribution (Top 5 Areas)

# Step 1: Identify top 5 areas by average wage
top_areas = (
    df.groupby('area_name')['annual_mean_wage']
    .mean()
    .sort_values(ascending=False)
    .head(5)
    .index
)
filtered_df = df[df['area_name'].isin(top_areas)]
plt.figure(figsize=(12,5))

sns.boxplot(
    x='area_name',
    y='annual_mean_wage',
    data=filtered_df
)

plt.yscale('log')
plt.title("Wage Distribution Across Top 5 Areas (Log Scale)")
plt.xlabel("Area")
plt.ylabel("Annual Mean Wage (Log Scale)")
plt.xticks(rotation=30, ha='right')
plt.show()


```

**Task 4: Documentation & Insights**

 Key Insights and Observations
 1. **Wage distribution is highly right-skewed**:  The distribution of annual mean wages is highly right-skewed, indicating that while most occupations earn moderate wages, a small number of roles command extremely high salaries.

2. **Wage levels are evenly distributed by design**:  Wage levels created using quantile-based classification are evenly distributed by design, enabling fair comparison of low, medium, high, and very high wage groups across years and occupations.

3. **Occupation category plays a major role in wage differences**: Occupation categories such as Engineering, Management, and Research consistently show higher median wages compared to Support and Clerical roles.

4. **Employment distribution is also right-skewed**: The number of employed individuals across occupations shows a similar right-skewed pattern. A few occupation categories employ a disproportionately large share of the workforce, while most occupations have relatively smaller employment counts.

5. **High employment does not necessarily mean high wages**: There is no strong linear relationship between the number of employed individuals and wage levels, suggesting that high employment does not necessarily correspond to higher wages.

6. **Wage distributions vary over time**- When analyzed by year, wage distributions show gradual upward movement, especially within higher wage categories. This suggests long-term wage growth trends, although disparities across occupations remain.
